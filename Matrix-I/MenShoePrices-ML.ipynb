```
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```
df_orig = pd.read_csv('Data/MenShoePrices.csv', low_memory=False)
```

```
df_orig.shape
```

```
y_true = df_orig['prices_amountmin']
mean_price_log = np.expm1(np.mean(np.log1p(y_true)))
y_pred = [mean_price_log] * y_true.shape[0]
mean_absolute_error(y_true, y_pred)
```

```
mean_price_log
```

```
df_orig.columns
```

```
df_orig['brand'].value_counts()
```

```
df_orig['brand_id'] = df_orig['brand'].factorize()[0]
```

```
X = df_orig[['brand_id']].values
y = df_orig['prices_amountmin'].values
```

```
model = DecisionTreeRegressor(max_depth=5)
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error')
np.mean(scores), np.std(scores)
```

```
def run_model(feature):
  X = df_orig[feature].values
  y = df_orig['prices_amountmin'].values
  model = DecisionTreeRegressor(max_depth=5)
  scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error')
  return np.mean(scores), np.std(scores)
```

```
run_model(['brand_id'])
```

```
df_orig['manufacturer'].value_counts()
```

```
df_orig['manufacturer_id'] = df_orig['manufacturer'].factorize()[0]
```

```
run_model(['manufacturer_id'])
```

```
run_model(['brand_id', 'manufacturer_id'])
```
